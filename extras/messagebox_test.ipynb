```python
import tkinter
import tkinter as tk
from tkinter import *
from tkinter import ttk
from tkinter import messagebox

def test_messagebox(win):
        messagebox.showinfo("Test", "Quiero que este mensaje no este ligado a la main window", parent=win)
```

```python
class EventWindow():
    def __init__(self,main_window): #Creamos un constructor que recibe objeto, ya que estamos declarando el main_window en toplevel
        self.win=main_window.win #A la variable self.win recibe del objeto solo win
        self.main_window=main_window #self.obj recibe el objeto entero, en esta linea recibe el objeto main_window con todas las funciones miembro
        self.event_window=tkinter.Tk()
        self.event_window.title("Eventos")
        self.event_window.configure(bg="#D49FFF")
        self.event_window.geometry("675x500")
        self.menuEvent_window()
    
        
    
    def menuEvent_window(self):
        new_menu= Menu(self.event_window)
        self.event_window.config(menu=new_menu)

        event_menu= Menu(new_menu, tearoff=0)#, tearoff=0
        new_menu.add_cascade(label= "Archivo", menu=event_menu)
        event_menu.add_command(label= "Salir...", command= self.event_window.destroy)

        help_menu= Menu(new_menu, tearoff=0) #, tearoff=0
        new_menu.add_cascade(label="Ayuda", menu= help_menu)
        help_menu.add_command(label= "Â¿Como mostrar un showinfo sin ligar a la main window?", command=lambda:test_messagebox(self.event_window))

  

    

        
```

```python
class MainWindow():
    def __init__(self):
        self.win= tkinter.Tk()
        self.win.title("BITACORA DE SISTEMAS")
        self.win.configure(bg="#D49FFF")
        self.win.geometry("1000x300")
        self.win.add_event=tk.Button(text="Nueva ventana", command= lambda:EventWindow(self)) #Antes self.win pero de esta forma solo pasaba la ventana a EventWindow. La forma correcta es solo self para pasar el objeto completo.
        self.win.add_event.pack()
        self.win.mainloop()
        

MainWindow()
```

```python
class Int():
    def __init__(self, i):
        self.int=i

    def __int__(self):
        return self.i
    def __abs__(self):
        return self.i.__abs__()
    def __add__(self,obj):
        return self.i+int(obj)
    def __and__(self,obj):
        return self.i&int(obj)
    def __bool__(self):
        return self.i.__bool__()
    def __ceil__(self):
        return self.i.__ceil__()
    def __class__(self):
        return self.i.__class__()
    def __delattr__(self,a):
        return self.i.__delattr__(a)
    def __divmod__(self):
        return self.i.__divmod__()
    def __doc__(self):
        return self.i.__doc__()
    def __eq__(self,obj):
        return self.i==int(obj)
    def __float__(self):
        return self.i.__float__()
    def __floor__(self):
        return self.i.__floor__()
    def __floordiv__(self):
        return self.i.__floordiv__()
    def __format__(self):
        return self.i.__format__()
    def __ge__(self,obj):
        return self.i>=int(obj)
    def __getattr__(self,a):
        return self.i.__getattr__(a)

dir(5)
```
